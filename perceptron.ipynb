```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("placement-dataset.csv")
```

```python
df.head()
```

```python
df=df.iloc[:,1:]
```

```python
df.head()
```

```python

from sklearn.preprocessing import StandardScaler
```

```python
X=df.iloc[:,:2]
Y=df.iloc[:,-1]
```

```python
scaler=StandardScaler()
scaler.fit(X)
xscaled=scaler.transform(X)
```

```python
xscaled=pd.DataFrame(xscaled,columns=X.columns)
```

```python
from sklearn.linear_model import Perceptron
```

```python
perc=Perceptron()
```

```python
perc.fit(xscaled,Y)
```

```python
perc.coef_
```

```python
perc.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(xscaled.values,Y.values,clf=perc, legend=2)
```
